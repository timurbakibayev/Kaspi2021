```python
cars = []
for i in range(1,5):
    result = requests.get("https://kolesa.kz/cars/toyota/almaty/?_sys-hasphoto=2&auto-custom=2&page="+str(i))    
    page2 = result.content.decode("UTF-8")
    page = BeautifulSoup(page2)
    for img in page.find_all("img"):
        alt = img["alt"]
        if "фото" not in alt and "toyota" in alt.lower():
            cars.append(alt)
```

```python
model = []
price = []
city = []
for car in cars:
    tg = car.find(" тг.")
    za = car.find(" за ")
    v = car.find(" в ")
    model.append(car[:za])
    price.append(int(car[za+4:tg].replace(" ","")))
    city.append(car[v+3:])
```

```python
df = pd.DataFrame({
    "model": model,
    "price": price,
    "city": city,
})
```
